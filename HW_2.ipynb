### Практическое задание 2.

**Вопрос 1**
В чём различие между зависимыми и независимыми выборками?

Если одному случаю из выборки X соответствует один и только один случай из выборки Y и наоборот,то такие выборки называются зависимыми. В случае, если такая взаимосвязь между выборками отсутствует, то эти выборки считаются независимыми. Соответственно, зависимые выборки всегда имеют одинаковый объём, а объём независимых может отличаться.

**Вопрос 2**
Когда применяются параметрические статистические критерии, а когда их непараметрические аналоги?

Параметрические критерии применяются, когда есть информация о характере распредления величины и достаточный объем выборки. Когда нет предоложений об характере распределения или выборка очень малая(грубо говоря данные низкого качества), то применяется непараметричесий аналог статистического критерия.
